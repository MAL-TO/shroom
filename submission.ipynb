```python
from pathlib import Path

BASE_DIR = Path("/data1/malto/shroom/")
```

```python
import pandas as pd

df = pd.read_csv(BASE_DIR / "sequential.csv")
df
```

```python
df3 = pd.read_csv(BASE_DIR / "baseline.csv")
df3
```

```python
df = df['sequential_nogpt_deberta_xlarge'] + df3['baseline']
```

```python
df
```

```python
df = df / 2
```

```python
df = df.to_frame()
```

```python
df
```

```python
df = df.reset_index()
```

```python
import json 
with open(BASE_DIR / f"test.model-agnostic.json") as f:
    j = json.loads(f.read())
```

```python
ids = [el['id'] for el in j]
```

```python
preds = []
for id in df['index']:
    prob = df[0][id]
    preds.append({'id': ids[id], 'p(Hallucination)': prob, 'label': "Hallucination" if prob > 0.5 else "Not Hallucination"})
```

```python
preds
```

```python
import json

with open(BASE_DIR / f"ensemble.json", "w", encoding="utf-8") as f:
    f.write(json.dumps(preds))
```
